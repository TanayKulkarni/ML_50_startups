```python

# coding: utf-8

# In[52]:

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# In[53]:

dataset = pd.read_csv('C:/Users/TANMAY/Desktop/50_Startups.csv')
x=dataset.iloc[:,0:4].values
y=dataset.iloc[:,4:].values
#print(x,"\n",y)


# In[54]:

from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# label encoder is used to convert categorical to numeric
# will assign converted (categorical to numeric) value into column
labelencoder_x_1 = LabelEncoder()
x[:, 3] = labelencoder_x_1.fit_transform(x[:, 3])
#print(pd.DataFrame(x))


# print(pd.DataFrame(x))

onehotencoder = OneHotEncoder(categorical_features = [3])
x = onehotencoder.fit_transform(x).toarray()
#print(pd.DataFrame(x))

x=x[:, 1:]



# In[55]:

from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

x=sc_x.fit_transform(x)
#print(pd.DataFrame(x))
y=sc_y.fit_transform(y)
#print(pd.DataFrame(y))




# splitting dataset into training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)#

# In[56]:
from sklearn.svm import SVR
regressor = SVR(kernel='linear')
regressor.fit(x_train,y_train)


# In[57]:

y_pred = regressor.predict(x_test)
sc_y.inverse_transform(y_pred)
print(pd.DataFrame(sc_y.inverse_transform(y_pred)))

print(pd.DataFrame(sc_y.inverse_transform(y_test)))


# In[58]:

from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(x_train,y_train)


# In[60]:

y_pred = regressor.predict(x_test)
sc_y.inverse_transform(y_pred)
print(pd.DataFrame(sc_y.inverse_transform(y_pred)))

print(pd.DataFrame(sc_y.inverse_transform(y_test)))


# In[61]:

y_pred = sc_y.inverse_transform(
    regressor.predict(
        sc_x.transform(
            np.array([[1,0,56445,87961,53131]])
        )
    )
)
print(y_pred)

```
